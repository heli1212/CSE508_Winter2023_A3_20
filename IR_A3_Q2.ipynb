```python
#!pip install networkx
import networkx as nwx
import pandas as pd
import matplotlib.pyplot as plt
```

```python
########### Read text file (Edge list) using networkx #############
dirGraph = nwx.read_edgelist('converted.txt', create_using=nwx.DiGraph())
df = pd.DataFrame(dirGraph)
df
```

```python
########## Directed Graph Information ######################
nodes = nwx.number_of_nodes(dirGraph)
edges = nwx.number_of_edges(dirGraph)

print(f"There are {nodes} nodes and {edges} edges in our dataset")
```

```python
pageRankScores = nwx.pagerank(dirGraph, alpha=0.85, tol=0.00009)
pageRank_list=[]
for node in pageRankScores:
    pageRank_list.append(pageRankScores[node])
#pageRank_list
pageRankScores
```

```python
########### Hits Algo to calculate authority and hub scores ###############

noOfHits = nwx.hits(dirGraph, tol=0.00009, nstart=None)
noOfHits
```

```python
authorityScore = noOfHits[1]
authorityScore_list=[]
print(authorityScore)
for node in authorityScore:
    authorityScore_list.append(authorityScore[node])
print(authorityScore_list)
```

```python
hubScore = noOfHits[0]
hubScore_list=[]
print(hubScore)
for node in hubScore:
    hubScore_list.append(hubScore[node])
print(hubScore_list)
```

```python
######################## COMPAIRISION #############################

#----------------PAGERANK-----------------------#
dictPageRank = dict(sorted(pageRankScores.items(), key=lambda item: item[1], reverse=True))
for i, (key, value) in enumerate(dictPageRank.items()):
    if i >= 10:
        break
    print(f"{key}: {value}")
```

```python
#----------------PAGERANK PLOT-----------------------#
plt.bar(range(len(pageRank_list)), pageRank_list, align='center', alpha=1,  label= "Indegree Plot")
plt.xlabel('Nodes')
plt.ylabel('Page Ranks')
plt.title('Page Rank Plot')
```

```python
#----------------Authority-----------------------#
dictAuthority = dict(sorted(authorityScore.items(), key=lambda item: item[1], reverse=True))
for i, (key, value) in enumerate(dictAuthority.items()):
    if i >= 10:
        break
    print(f"{key}: {value}")
    
```

```python
#---------------- Authority PLOT-----------------------#
plt.bar(range(len(authorityScore_list)), authorityScore_list, align='center', alpha=1,  label= "Indegree Plot")
plt.xlabel('Nodes')
plt.ylabel('Authority')
plt.title('Authority Plot')
```

```python
#----------------Hub-----------------------#
dictHub = dict(sorted(hubScore.items(), key=lambda item: item[1], reverse=True))
for i, (key, value) in enumerate(dictHub.items()):
    if i >= 10:
        break
    print(f"{key}: {value}")
```

```python
#---------------- Hub PLOT-----------------------#
plt.bar(range(len(hubScore_list)), hubScore_list, align='center', alpha=1,  label= "Indegree Plot")
plt.xlabel('Nodes')
plt.ylabel('Hub')
plt.title('Hub Plot')
```

```python
##################################### Combined Comparision #######################################
# create sample data
index_arr = range(1005)
col1_arr = pageRank_list
col2_arr = authorityScore_list
col3_arr = hubScore_list

# create DataFrame with one array as index and three arrays as columns
df = pd.DataFrame({'PageRank': col1_arr, 'Authority': col2_arr, 'Hub': col3_arr}, index=index_arr)
df.index.name = 'Node'

# print DataFrame
print(df)
```
